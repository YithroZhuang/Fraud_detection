```python
partners = df['partnerid'].unique().tolist()
offers = df['cid'].unique().tolist()
clients = df['client'].unique().tolist()

with open('/data/zy/youmi/FDMA/metapath/id_partner.txt', 'w') as f:
    for ind, cc in enumerate(partners):
        f.write(str(ind) + '\t' + cc + '\n')

with open('/data/zy/youmi/FDMA/metapath/id_offer.txt', 'w') as f:
    for ind, cc in enumerate(offers):
        f.write(str(ind) + '\t' + cc + '\n')
        
with open('/data/zy/youmi/FDMA/metapath/id_client.txt', 'w') as f:
    for ind, cc in enumerate(clients):
        f.write(str(ind) + '\t' + str(cc) + '\n')
```

```python
clients[0]
```

```python
dd = {}
for cc in partners:
    dd[cc] = 'p'
for cc in offers:
    dd[cc] = 'o'
for cc in clients:
    dd[cc] = 'c'

des_f = open('/data/zy/youmi/FDMA/metapath/node_type_mapings.txt', 'w')
with open('/data/zy/youmi/FDMA/metapeth/random_walk.txt', 'r') as f:
    for line in f:
        splits = line.strip().split('\t')
        for split in splits:
            des_f.write(split + '\t' + dd[split] + '\n')

des_f.close()
```
