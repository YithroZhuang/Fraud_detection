```python
import pandas as pd
import numpy as np
```

```python
path = '/data/youmi/'
df = []
for i in range(1, 6):
    d = pd.read_csv(path + 'conversion-2018-05-part' + str(i) + '.csv', low_memory=False)
    # df.columns = d.columns.values.tolist()
    df.append(d)
df = pd.concat(df, axis=0)
    
df2 = pd.read_excel(path + 'fraud_transaction.xlsx', sheetname=None)
```

```python
users = df['dvid'].unique().tolist()
offers = df['oid_'].unique().tolist()
traffic = df['tid'].unique().tolist()

with open('/data/zy/youmi/test_data/id_users.txt', 'w') as f:
    for ind, cc in enumerate(users):
        f.write(str(ind) + '\t' + str(cc) + '\n')

with open('/data/zy/youmi/test_data/id_offers.txt','w') as f:
    for ind, cc in enumerate(offers):
        f.write(str(ind) + '\t' + str(cc) + '\n')

with open('/data/zy/youmi/test_data/id_traffic.txt', 'w') as f:
    for ind, cc in enumerate(traffic):
        f.write(str(ind) + '\t' + str(cc) + '\n')
```

```python
traffic_user, traffic_offer = [] , []
for ind in df.index:
    traffic_user.append((df.loc[ind, 'tid'], df.loc[ind, 'dvid']))
    traffic_offer.append((df.loc[ind, 'tid'], df.loc[ind, 'oid_']))
    
with open('/data/zy/youmi/test_data/traffic_users.txt', 'w') as f:
    for i in range(len(traffic_user)):
        f.write(str(traffic_user[i][0] + '\t' + str(traffic_user[i][1]) + '\n'))

with open('/data/zy/youmi/test_data/traffic_offer.txt', 'w') as f:
    for i in range(len(traffic_offer)):
        f.write(str(traffic_offer[i][0] + '\t' + str(traffic_offer[i][1]) + '\n'))
```
